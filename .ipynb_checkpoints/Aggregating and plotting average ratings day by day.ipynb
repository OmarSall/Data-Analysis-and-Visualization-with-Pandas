```python
import pandas as pd
from datetime import datetime
from pytz import utc
import matplotlib.pyplot as plt

data = pd.read_csv("reviews.csv", parse_dates = ['Timestamp'])
data.head()
#data.shape()
```

## Rating average/count by day

```python
#adding a new column - day, dt is a property that gives access to datatypes
data['Day'] = data['Timestamp'].dt.date

day_average = data.groupby(['Day']).mean()      # .mean is a mothod of aggregation
day_average
```

```python
day_average = data.groupby(['Day']).count()
```

```python
day_average.index
```

```python
day_average['Rating']
```

```python
plt.figure(figsize = (25, 3))
plt.plot(day_average.index, day_average['Rating'])
```

```python
day_average = data.groupby(['Day']).count()
day_average.head()
```

## Rating average by week

```python
data
```

```python
#strf - string from time
data['Week'] = data['Timestamp'].dt.strftime('%Y-%U')
data
```

```python
week_average = data.groupby(['Week']).mean()
week_average
```

```python
#week column is an index!

week_average.index
```

```python
plt.figure(figsize = (25, 3))
plt.plot(week_average.index, week_average['Rating'])
```

### Average rating by month by course

```python
data['Month'] = data['Timestamp'].dt.strftime('%Y-%m')
data.head()
```

```python
month_average_crs = data.groupby(['Month', 'Course Name'])['Rating'].mean().unstack()
month_average_crs.index
#without unstack() we would have two indexes, unstack gives better structure
```

```python
#month_average_crs
month_average_crs.columns
```

```python
month_average_crs[-20:]
```

```python
month_average_crs.plot(figsize=(25,10))
```

```python
month_average_crs = data.groupby(['Month', 'Course Name'])['Rating'].count().unstack()
month_average_crs.plot(figsize=(25,10))
```

### What day are people the happiest

```python
data
```

```python
data['Weekday'] = data['Timestamp'].dt.strftime('%A')
data['Daynumber'] = data['Timestamp'].dt.strftime('%w')

weekday_average = data.groupby(['Weekday', 'Daynumber']).mean()
weekday_average = weekday_average.sort_values('Daynumber')
weekday_average.index.get_level_values(0) #we are extracting first column of the muti-index
```

```python
plt.figure(figsize = [25, 3])
plt.plot(weekday_average.index.get_level_values(0), weekday_average['Rating'])
```
